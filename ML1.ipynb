```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
n=3
p=0.5
x=list(range(0,n+1))
```

```python
x
```

```python
y = stats.binom.pmf(x,n,p)
z = stats.poisson.pmf(x,n*p)
```

```python
y
```

```python
z
```

```python
stats.binom.pmf(3,n,p)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.ylabel("Probability")
plt.xlabel("RV X # of Heads")
```
